# Plots of Bijections for eyeballing

```python
%autoreload 2
import jax.numpy as np
import pylab as pl
from jaxrk.core.constraints import SoftBound, CholeskyBijection, Softplus, SquarePlus, SquareSquash
```

```python
x = np.linspace(-50,50, 1000)
nneg = np.linspace(1e-5,20, 1000)
interv = np.linspace(1e-5, 1 - 1e-5, 500)

for bij, insp, outsp in [(Softplus(), x, nneg),
                            (SquarePlus(), x, nneg),
                            (SquareSquash(), x, nneg),]:
    pl.figure()
    f = bij(insp)

    pl.plot(bij.inv(f), f, "-", label=f"{bij.__class__.__name__} - inverse")
    pl.plot(insp, f, ":", label=f"{bij.__class__.__name__} - forward")
    pl.legend(loc = "best")
```
